```python
import pandas as pd
```

```python
excel_file = 'movies.xlsx'
movies = pd.read_excel(excel_file)
```

```python
movies.head()
```

```python
# read a specific sheet
movies_sheet1 = pd.read_excel(excel_file, sheet_name='2010')
movies_sheet1.head()
```

```python
# Read all the sheets in an Excel file
xlsx = pd.ExcelFile(excel_file)
movies_sheets = []
for sheet in xlsx.sheet_names:
    movies_sheets.append(xlsx.parse(sheet))
movies = pd.concat(movies_sheets)
movies.info()
```

```python
# Read a subset of columns
movies_subset_columns = pd.read_excel(excel_file, sheet_name='2010', usecols=[0,1,2,3,4,9])
movies_subset_columns.head()
```

```python
# Use pivot table 
movies_subset = movies_subset_columns[['Year', 'Gross Earnings']]
earnings_by_year = movies_subset.pivot_table(index=['Year'])
earnings_by_year.head()
```

```python
# Write to an Excel file
writer = pd.ExcelWriter('output.xlsx', engine='xlsxwriter')
movies.to_excel(writer, index=False, sheet_name='report')
workbook = writer.book
worksheet = writer.sheets['report']
row_fmt = workbook.add_format({'bold': False, 'italic': True})
worksheet.set_row(1, None, row_fmt)
writer.save()
```

```python
writer = pd.ExcelWriter('output2.xlsx', engine='xlsxwriter')
earnings_by_year.to_excel(writer, sheet_name='GrossEarnings')
workbook = writer.book
worksheet = writer.sheets['GrossEarnings']
writer.save()
```
